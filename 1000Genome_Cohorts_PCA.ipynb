```r
# Import subsample of 1000 genomes cohort ancestry data
name_of_file_in_bucket <- '1000_genomes_subset.csv'

########################################################################
##
################# DON'T CHANGE FROM HERE ###############################
##
########################################################################

# Copy the file from current workspace to the bucket
system(paste0("gsutil cp ", my_bucket, "/data/", name_of_file_in_bucket, " ."), intern=T)

# Load the file into a dataframe
simple_ancestry_df  <- read_csv(name_of_file_in_bucket)
#simple_ancestry_df <-as.tibble(simple_ancestry_df)
simple_ancestry_df$Cohort <- "1000 Genomes"
head(simple_ancestry_df)
```

```r
#Subset T2D dataset, so only id column (person_id) and population columns remain
t2d_ancestry_df <- t2d_ancestry_df[,c(1,9:15)]
head(t2d_ancestry_df)
```

```r
#Subset A1C dataset, so only id column (person_id) and population columns remain
a1c_ancestry_df <- a1c_ancestry_df[,c(1,10:16)]
head(a1c_ancestry_df)
```

```r
#Concatentate T2D and 1000 genomes datasets into one 
t2d_1000_df <- rbind(simple_ancestry_df, t2d_ancestry_df)
head(t2d_1000_df)
```

```r
#Concatentate A1C and 1000 genomes datasets into one 
a1c_1000_df <- rbind(simple_ancestry_df, a1c_ancestry_df)
head(a1c_1000_df)
```

```r
#Convert person_id column in row names
test.t2d <- column_to_rownames(t2d_1000_df, var = "person_id")
t2d_ancestry_data <- test.t2d[,1:6]
head(t2d_ancestry_data)
```

```r
#Convert person_id column in row names
test.a1c <- column_to_rownames(a1c_1000_df, var = "person_id")
a1c_ancestry_data <- test.a1c[,1:6]
head(a1c_ancestry_data)
```

```r
#Normalize the T2D and 1000 combined dataset

t2d_ancestry_data_norm <- scale(t2d_ancestry_data)
#head(t2d_ancestry_data_norm[,-1])
head(t2d_ancestry_data_norm)
```

```r
#Calculate the principle components
t2d.pca <- princomp(t2d_ancestry_data_norm,cor=TRUE)
summary(t2d.pca, loadings=TRUE)
```

```r
#This is check the loadings
t2d.pca$loadings[,1:2]
```

```r
head(scale(t2d_ancestry_data_norm, center = t2d.pca$center, scale = t2d.pca$scale) %*% t2d.pca$loadings)
```

```r
head(t2d.pca$scores)
```

```r
t2d.scores <- data.frame(t2d.pca$scores)
head(t2d.scores)
```

```r
$#Formating the metadata associated with the T2D+1000 dataset so we can identify which is which when plottig
t2d_format_meta <- t2d_1000_df[,c("person_id","Cohort","European","African")]
#simple_format_meta <- simple_ancestry_df[,c("SampleID","Cohort")]
#combined_format_meta <- rbind(simple_format_meta, t2d_format_meta)
t2d.scores.meta <- cbind(t2d_format_meta, t2d.scores)
head(t2d.scores.meta)
head(t2d_format_meta)
```

```r
#Plotting the PCA for T2D + 1000 Genomes dataset
p1 <- ggplot(t2d.scores.meta) + 
      geom_point(aes(x = Comp.1, y = Comp.2,colour=Cohort), alpha = 0.3)
p1
```

```r
t2d_pca_loadings <- loadings(t2d.pca)[]
t2d_pca_importance <- summary(t2d.pca)$importance[2,]
plot.t2d.loadings <- data.frame(t2d_pca_loadings)
#plot.t2d.loadings
summary(t2d.pca)
PoV <- t2d.pca$sdev^2/sum(t2d.pca$sdev^2)
```

```r
#Adding labels and percent variance of components
p2 <- p1 + 
  xlab(paste("PC1 (", round(PoV[1]*100, 1),"%)", sep = "")) + 
  ylab(paste("PC2 (", round(PoV[2]*100, 1),"%)", sep = "")) +
  ggtitle("T2D Participants and !000 Genomes Subset")

p2
```

```r
#Adding population biplot on top
p3 <- p2 + 
  geom_segment(data = plot.t2d.loadings, aes(x = 0, y = 0, xend = Comp.1, yend = Comp.2), arrow = arrow(length = unit(0.03, "npc")), alpha = 0.2) + 
  geom_text(data = plot.t2d.loadings, mapping = aes(x = Comp.1, y = Comp.2, label = rownames(plot.t2d.loadings)), hjust = 1, vjust = -0.2, colour = "darkred", size = 4, check_overlap = TRUE)
p3
```

```r
#Similar approach to before on this time, colored by % European genetic ancestry
p1.grad <- ggplot(t2d.scores.meta) + 
           geom_point(aes(x = Comp.1, y = Comp.2,colour=European), alpha = 0.3) +
           scale_color_gradientn(colours = viridis(5))

p1.grad
```

```r
p2.grad <- p1.grad + 
  xlab(paste("PC1 (", round(PoV[1]*100, 1),"%)", sep = "")) + 
  ylab(paste("PC2 (", round(PoV[2]*100, 1),"%)", sep = "")) +
  ggtitle("T2D Participants and !000 Genomes Subset")

p2.grad
```
